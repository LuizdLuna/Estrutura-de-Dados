```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self._size = 0

    def addFirst(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = self.tail = new_node
        else:
            new_node.next = self.head
            self.head = new_node
        self._size += 1

    def addLast(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self._size += 1

    def first(self):
        return self.head.value if self.head else None

    def last(self):
        return self.tail.value if self.tail else None

    def removeFirst(self):
        if self.head is None:
            return None
        removed_value = self.head.value
        self.head = self.head.next
        if self.head is None:
            self.tail = None
        self._size -= 1
        return removed_value

    def size(self):
        return self._size

    def isEmpty(self):
        return self._size == 0

    def display(self):
        current = self.head
        elements = []
        while current:
            elements.append(current.value)
            current = current.next
        return elements

linked_list = SinglyLinkedList()

linked_list.addFirst(10)
linked_list.addFirst(20)
linked_list.addFirst(30)
print("Lista após addFirst:", linked_list.display())

linked_list.addLast(40)
linked_list.addLast(50)
print("Lista após addLast:", linked_list.display())  

print("Primeiro elemento:", linked_list.first())  
print("Último elemento:", linked_list.last())    

print("Elemento removido:", linked_list.removeFirst())
print("Lista após removeFirst:", linked_list.display())  

print("Tamanho da lista:", linked_list.size())

print("A lista está vazia?:", linked_list.isEmpty())

while not linked_list.isEmpty():
    print("Removendo:", linked_list.removeFirst())

print("A lista está vazia após remoções?:", linked_list.isEmpty()) 
```
