#**Robustez**
#####Um exemplo de robustez em Python pode ser um método que lida com entradas inesperadas de forma simples, como tratar exceções

````python
class Calculator:
    def divide(self, a, b):
        # Tenta executar a operação dentro do bloco try
        try:
            # Tenta dividir 'a' por 'b' e armazenar o resultado
            result = a / b
        # Captura a exceção ZeroDivisionError  quando 'b' é zero
        except ZeroDivisionError:
            # Define a mensagem de erro  para divisão por zero
            result = "Erro: Divisão por zero não é permitida."
        # Captura a exceção específica TypeError que ocorre quando os tipos de 'a' ou 'b' não são números (int ou float)
        except TypeError:
            # Define a mensagem de erro para um tipo de entrada inválida
            result = "Erro: Entrada inválida. Certifique-se de que ambos os números sejam inteiros ou floats."
        # Retorna o resultado da divisão ou a mensagem de erro
        return result
````

try: O bloco try é usado para envolver o código que pode potencialmente gerar uma exceção. Se o código dentro do bloco try executar sem erros, o programa continuará normalmente. Mas se uma exceção ocorrer, o fluxo de execução será imediatamente transferido para o bloco except correspondente.

except: O bloco except captura e trata exceções que ocorrem no bloco try. Neste código, há dois blocos except:

-->**except ZeroDivisionError**: Se o usuário tentar dividir por zero, o programa captura essa exceção e retorna uma mensagem de erro clara, em vez de travar.

-->**except TypeError**: Se o usuário passar valores que não são números (por exemplo, strings ou listas), o programa captura a exceção e retorna uma mensagem de erro informativa. Tipos esperado (int ou float).

#Vamos verificar **"robustez"** em java!!!

```java
import java.util.InputMismatchException;
import java.util.Scanner;

public class Calculator {

    // Método para dividir dois números
    public double divide(double a, double b) {
        if (b == 0) {
            throw new ArithmeticException("Erro: Divisão por zero não é permitida.");
        }
        return a / b;
    }

    public static void main(String[] args) {
        Calculator calculator = new Calculator();
        Scanner scanner = new Scanner(System.in);

        try {
            // Solicita ao usuário os valores de entrada
            System.out.print("Digite o primeiro número (a): ");
            double a = scanner.nextDouble();

            System.out.print("Digite o segundo número (b): ");
            double b = scanner.nextDouble();

            // Realiza a divisão
            double result = calculator.divide(a, b);
            System.out.println("Resultado da divisão: " + result);

        } catch (InputMismatchException e) {
            // Captura exceção se a entrada não for um número válido
            System.out.println("Erro: Entrada inválida. Certifique-se de digitar números.");
        } catch (ArithmeticException e) {
            // Captura exceção de divisão por zero
            System.out.println(e.getMessage());
        } finally {
            // Fecha o scanner para evitar vazamentos de recursos
            scanner.close();
        }
    }
}
```

# Comparação de Robustez: Python vs Java

## Similaridades

### Tratamento de Exceções

- Tanto Python quanto Java suportam o uso de blocos `try-catch` (em Java) e `try-except` (em Python) para capturar e tratar exceções.
- Ambos permitem capturar exceções específicas e fornecer feedback claro ao usuário.

**Exemplo em Python**:
```python
try:
    result = a / b
except ZeroDivisionError:
    print("Erro: Divisão por zero não é permitida.")
```

**Exemplo em Java**:

```java
try {
    result = a / b;
} catch (ArithmeticException e) {
    System.out.println("Erro: Divisão por zero não é permitida.");
}
```
### Validação de Entrada:

- Ambos permitem a validação de entradas para garantir que os dados sejam do tipo esperado antes de processá-los.

- Em ambos, é possível usar estruturas condicionais (if-else) para verificar tipos ou valores.

### Boas Práticas:

- Em Python e Java, é recomendado fechar recursos (como arquivos ou conexões) após o uso, seja com finally (Java) ou with (Python).

### Feedback ao Usuário:

- Ambos permitem fornecer mensagens de erro claras e informativas para ajudar o usuário a entender o que deu errado.

## Diferenças

### Tipagem:

- Java: É uma linguagem fortemente tipada. O compilador verifica os tipos em tempo de compilação, o que ajuda a prevenir erros de tipo antes mesmo de executar o código.

  - Exemplo: Se você tentar dividir uma String por um int, o código nem compila.

- Python: É uma linguagem dinamicamente tipada. Os tipos são verificados em tempo de execução, o que pode levar a erros durante a execução se os tipos forem incompatíveis.

- Exemplo: Se você tentar dividir uma string por um int, o código só falhará quando a linha for executada.

### Exceções Verificadas vs Não Verificadas:

- Java: Possui exceções verificadas (checked exceptions), que devem ser explicitamente tratadas ou declaradas no método (com throws). Isso força o desenvolvedor a lidar com possíveis erros.

  - Exemplo: IOException em operações de leitura/escrita de arquivos.

- Python: Não possui exceções verificadas. Todas as exceções são não verificadas (unchecked), o que significa que o tratamento é opcional.

### Sintaxe e Simplicidade:

- Python: A sintaxe é mais simples e direta, o que facilita a escrita de código robusto rapidamente. Por exemplo, o tratamento de exceções é menos verboso.

- Java: A sintaxe é mais verbosa, o que pode tornar o código mais robusto, mas também mais complexo de escrever e manter.

### Gerenciamento de Recursos:

- Python: Usa o padrão with para gerenciar recursos (como arquivos), garantindo que eles sejam fechados automaticamente.

  - Exemplo:
  ```python
  with open("arquivo.txt", "r") as file:
    content = file.read()
    ```
- Java: Usa blocos try-with-resources para garantir o fechamento automático de recursos.    
  - Exemplo:
  ```java
try (BufferedReader br = new BufferedReader(new FileReader("arquivo.txt"))) {
    String line = br.readLine();
}
  ```

### Cultura de Desenvolvimento:

- Python: A filosofia "EAFP" (É mais fácil pedir perdão do que permissão) é comum, incentivando o uso de try-except para lidar com erros.

- Java: A filosofia "LBYL" (Look Before You Leap) é mais comum, incentivando a verificação de condições antes de executar operações.

#Adaptabilidade

##### Adaptabilidade pode ser demonstrada com a capacidade de um objeto se adaptar a diferentes necessidades, como mudar a forma de calcular descontos dependendo do tipo de cliente



````python
class DiscountStrategy:
    def get_discount(self, amount):
        return 0  # Estratégia padrão que não aplica desconto

class RegularCustomerDiscount(DiscountStrategy):
    def get_discount(self, amount):
        return amount * 0.1  # Aplica 10% de desconto para clientes regulares

class VIPCustomerDiscount(DiscountStrategy):
    def get_discount(self, amount):
        return amount * 0.2  # Aplica 20% de desconto para clientes VIP

class Billing:
    def __init__(self, discount_strategy):
        self.discount_strategy = discount_strategy

    def calculate_final_amount(self, amount):
        discount = self.discount_strategy.get_discount(amount)
        return amount - discount  # Calcula o valor final após aplicar o desconto
````

#Testando!

````python
# Exemplo de uso:
billing_regular = Billing(RegularCustomerDiscount())
print(billing_regular.calculate_final_amount(100))  # Saída: 90.0

billing_vip = Billing(VIPCustomerDiscount())
print(billing_vip.calculate_final_amount(100))  # Saída: 80.0
````

#Explicando!

#####DiscountStrategy é uma classe base com um método get_discount que retorna 0 por padrão.

#####RegularCustomerDiscount e VIPCustomerDiscount são subclasses que implementam suas próprias lógicas de desconto.

#####A classe Billing aceita uma estratégia de desconto como parâmetro no seu construtor.

#####O método calculate_final_amount calcula o valor final aplicando o desconto adequado.

#####Dessa forma, a classe Billing é adaptável a diferentes estratégias de desconto sem precisar ser modificada diretamente.

#Vamos verificar **"adaptabilidade"** em java!!!
```java
// Interface que define a estratégia de desconto
interface DiscountStrategy {
    double getDiscount(double amount);
}

// Estratégia padrão que não aplica desconto
class NoDiscount implements DiscountStrategy {
    @Override
    public double getDiscount(double amount) {
        return 0; // Sem desconto
    }
}

// Estratégia de desconto para clientes regulares
class RegularCustomerDiscount implements DiscountStrategy {
    @Override
    public double getDiscount(double amount) {
        return amount * 0.1; // 10% de desconto
    }
}

// Estratégia de desconto para clientes VIP
class VIPCustomerDiscount implements DiscountStrategy {
    @Override
    public double getDiscount(double amount) {
        return amount * 0.2; // 20% de desconto
    }
}

// Classe que utiliza a estratégia de desconto
class Billing {
    private DiscountStrategy discountStrategy;

    // Construtor que recebe a estratégia de desconto
    public Billing(DiscountStrategy discountStrategy) {
        this.discountStrategy = discountStrategy;
    }

    // Método para calcular o valor final após o desconto
    public double calculateFinalAmount(double amount) {
        double discount = discountStrategy.getDiscount(amount);
        return amount - discount;
    }
}

// Exemplo de uso
public class Main {
    public static void main(String[] args) {
        // Estratégia padrão (sem desconto)
        DiscountStrategy noDiscount = new NoDiscount();
        Billing billing = new Billing(noDiscount);
        System.out.println("Valor final (sem desconto): " + billing.calculateFinalAmount(100)); // Saída: 100.0

        // Estratégia de desconto para clientes regulares
        DiscountStrategy regularDiscount = new RegularCustomerDiscount();
        billing = new Billing(regularDiscount);
        System.out.println("Valor final (desconto regular): " + billing.calculateFinalAmount(100)); // Saída: 90.0

        // Estratégia de desconto para clientes VIP
        DiscountStrategy vipDiscount = new VIPCustomerDiscount();
        billing = new Billing(vipDiscount);
        System.out.println("Valor final (desconto VIP): " + billing.calculateFinalAmount(100)); // Saída: 80.0
    }
}
```

# Adaptabilidade em Python vs Java: Diferenças e Similaridades

Tanto Python quanto Java suportam a criação de sistemas adaptáveis, mas cada linguagem tem suas particularidades.

## Similaridades

### Uso de Interfaces/Protocolos

- Python: Em Python, a adaptabilidade é alcançada usando classes base ou protocolos (como o ABC ou o conceito de "duck typing"). No exemplo, a classe DiscountStrategy atua como uma interface implícita.

- Java: Java usa interfaces explicitamente para definir contratos. No exemplo, a interface DiscountStrategy define o método getDiscount, que todas as estratégias devem implementar.

### Baixo Acoplamento

- Em ambas as linguagens, a classe Billing depende de uma abstração (DiscountStrategy em Java ou a classe base em Python), e não de implementações concretas. Isso reduz o acoplamento e facilita a manutenção.

### Flexibilidade em Tempo de Execução

- Tanto Python quanto Java permitem que a estratégia seja alterada dinamicamente em tempo de execução. Por exemplo, você pode passar diferentes estratégias de desconto para a classe Billing conforme necessário.

## Diferenças

### Sintaxe e Verbosidade

- Python: A sintaxe é mais concisa e menos verbosa. Por exemplo, em Python, você não precisa declarar tipos explicitamente, e a herança é simples.

```python
class VIPCustomerDiscount(DiscountStrategy):
    def get_discount(self, amount):
        return amount * 0.2
```
- Java: Java é mais verboso, exigindo a declaração explícita de tipos, interfaces e modificadores de acesso.

```java
class VIPCustomerDiscount implements DiscountStrategy {
    @Override
    public double getDiscount(double amount) {
        return amount * 0.2;
    }
}
```
### Tipagem

- Python: Python é dinamicamente tipado, o que significa que os tipos sãoverificados em tempo de execução. Isso torna o código mais flexível, mas pode levar a erros em tempo de execução se o tipo esperado não for respeitado.

- Java: Java é estaticamente tipado, o que significa que os tipos são verificados em tempo de compilação. Isso aumenta a segurança do código, mas exige mais cuidado ao definir tipos e interfaces.

### Duck Typing vs Interfaces Explícitas

- Python: Python usa duck typing, onde o foco está no comportamento (se um objeto tem um método get_discount, ele pode ser usado como uma estratégia). Isso torna o código mais flexível, mas menos seguro em termos de tipos.

- Java: Java exige interfaces explícitas, o que torna o código mais seguro e previsível, mas menos flexível.

### Gerenciamento de Memória
- Python: Python usa coleta de lixo automática e não permite gerenciamento manual de memória. Isso simplifica o desenvolvimento, mas pode levar a vazamentos de memória em casos complexos.
- Java: Java também usa coleta de lixo, mas oferece mais controle sobre o ciclo de vida dos objetos (por exemplo, com finalize ou try-with-resources).




# Reutilização

##### Reutilização pode ser exemplificada criando-se componentes reutilizáveis, como classes que podem ser usadas em diferentes contextos.


````python
class Logger:
    def log(self, message):
        # Imprime a mensagem recebida no console, precedida por "Log: "
        print(f"Log: {message}")

class Application:
    def __init__(self, logger):
        # Inicializa a classe Application com um objeto Logger
        self.logger = logger

    def do_something(self):
        # Usa o logger para registrar uma mensagem específica
        self.logger.log("Ação executada na aplicação")

class AnotherApplication:
    def __init__(self, logger):
        # Inicializa a classe AnotherApplication com um objeto Logger
        self.logger = logger

    def perform_action(self):
        # Usa o logger para registrar uma mensagem específica
        self.logger.log("Ação executada em outra aplicação")
````

#Testando!

````python
# Logger pode ser reutilizado em diferentes aplicações
logger = Logger()

app1 = Application(logger)
app1.do_something()  # Saída: Log: Ação executada na aplicação

app2 = AnotherApplication(logger)
app2.perform_action()  # Saída: Log: Ação executada em outra aplicação
````

##Similaridades

    - Exemplo:

        Em Python: Application(logger)

        Em Java: Application app = new Application(logger);

### Injeção de Dependência

- Tanto Python quanto Java suportam a injeção de dependência, onde dependências (como objetos de outras classes) são passadas para uma classe via construtor ou métodos. Isso promove a reutilização e facilita a manutenção do código.

### Modularidade

- Ambas as linguagens incentivam a modularidade, onde funcionalidades são divididas em classes ou módulos independentes que podem ser reutilizados em diferentes partes do sistema.

## Diferenças
### Sintaxe e Verbosidade

- Python: A sintaxe é mais concisa e menos verbosa. Por exemplo, a criação de classes e métodos é mais simples e direta.

```python
class Logger:
    def log(self, message):
        print(f"Log: {message}")
```

- Java: A sintaxe é mais verbosa, exigindo a definição explícita de tipos, modificadores de acesso e blocos de código.

```java
public class Logger {

        System.out.println("Log: " + message);
    }

```

### Herança Múltipla

- Python: Suporta herança múltipla, onde uma classe pode herdar de várias classes base. Isso pode ser útil para reutilização, mas também pode levar a complexidade e conflitos.

```python
class A:
```

```java
class C implements A, B {  // Implementação de múltiplas interfaces
```

- Python: Utiliza módulos e pacotes para organizar e reutilizar código. A importação de módulos é simples e direta.
```python

```

- Java: Utiliza pacotes (packages) para organizar classes e interfaces. A importação de classes é mais explícita e requer o uso do caminho completo do pacote.

### Flexibilidade vs Rigidez


- Python: É mais flexível, permitindo a reutilização de código de forma dinâmica e com menos regras. Isso pode ser vantajoso para prototipagem rápida, mas pode levar a código menos estruturado.

- Java: É mais rígido, com regras claras para reutilização (como interfaces, classes abstratas e modificadores de acesso). Isso promove um código mais seguro e organizado, mas pode exigir mais esforço inicial.

#Modularidade
Modularidade refere-se à divisão de um sistema em partes menores e gerenciáveis (módulos). Cada módulo é responsável por uma funcionalidade específica.

Módulo de Livros (book.py):

```python
class Book:
    """Representa um livro na biblioteca."""
    
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_checked_out = False

    def check_out(self):
        if not self.is_checked_out:
            self.is_checked_out = True
            return True
        return False

    def return_book(self):
        if self.is_checked_out:
            self.is_checked_out = False
            return True
        return False
```

Módulo de Membros (member.py):

```python
class Member:
    """Representa um membro da biblioteca."""
    
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        self.checked_out_books = []

    def check_out_book(self, book):
        if book.check_out():
            self.checked_out_books.append(book)
            return True
        return False

    def return_book(self, book):
        if book in self.checked_out_books:
            if book.return_book():
                self.checked_out_books.remove(book)
                return True
        return False
```

#Testando!

````python
from book import Book
from member import Member

# Criando livros
book1 = Book('1984', 'George Orwell', '1234567890')
book2 = Book('To Kill a Mockingbird', 'Harper Lee', '0987654321')

# Criando um membro
member = Member('Alice', 'M001')

# Membro pega um livro emprestado
if member.check_out_book(book1):
    print(f"{member.name} pegou {book1.title} emprestado.")
else:
    print(f"{book1.title} não está disponível.")

# Membro retorna um livro
if member.return_book(book1):
    print(f"{member.name} retornou {book1.title}.")
else:
    print(f"{book1.title} não pode ser retornado.")
````

# Comparação entre Python e Java em Modularidade

### Tanto Python quanto Java suportam o conceito de modularidade, mas cada linguagem implementa essa ideia de maneiras diferentes.

## Similaridades

### Organização do Código

- Tanto Python quanto Java permitem organizar o código em módulos ou pacotes para facilitar a manutenção e a reutilização.

- Em Python, os módulos são arquivos .py, enquanto em Java, os módulos são pacotes (pastas) ou módulos do JPMS (Java Platform Module System).

###  Encapsulamento

-  Ambas as linguagens permitem encapsular funcionalidades relacionadas em módulos ou classes, expondo apenas o necessário para outros componentes.

###  Reutilização

- Em ambas as linguagens, módulos ou pacotes podem ser reutilizados em diferentes partes do programa ou até mesmo em outros projetos.

### Facilidade de Manutenção

- A modularidade em ambas as linguagens facilita a manutenção do código, pois alterações em um módulo não afetam outros, desde que as interfaces permaneçam consistentes.

## Diferenças

### Estrutura de Módulos

- Python: Módulos são simplesmente arquivos .py. Pacotes são diretórios que contêm um arquivo __init__.py. A modularidade é mais informal e flexível.

- Java: Módulos são organizados em pacotes (pastas) que refletem a estrutura de diretórios. A partir do Java 9, o JPMS permite criar módulos com dependências explícitas usando o arquivo module-info.java. A modularidade é mais rígida e estruturada.

### Importação de Módulos

- Python: A importação é feita com a palavra-chave import. Exemplo: import module ou from module import function. Não há necessidade de declarar dependências explicitamente.

- Java: A importação é feita com a palavra-chave import. Exemplo: import com.example.module.Class. No JPMS, as dependências devem ser declaradas explicitamente no arquivo module-info.java.

### Visibilidade e Controle de Acesso

- Python: Não há controle de acesso rígido. Tudo é público por padrão. Convenções como o uso de _ (underline) indicam que um atributo ou método é privado.

- Java: Possui controle de acesso explícito com modificadores como public, private, protected e package-private. O JPMS permite controlar quais pacotes são expostos para outros módulos.

### Sistema de Módulos

- Python: Não há um sistema de módulos nativo tão robusto quanto o JPMS. A modularidade é baseada em arquivos e diretórios.

- Java: O JPMS (Java 9+) introduziu um sistema de módulos nativo que permite definir dependências e encapsulamento em nível de módulo. Exemplo: module-info.java declara dependências e exportações.

### Flexibilidade vs. Rigidez

- Python: Mais flexível e dinâmico. Módulos podem ser importados e modificados em tempo de execução.

- Java: Mais rígido e estático. A estrutura de módulos e pacotes é verificada em tempo de compilação.

### Ferramentas de Gerenciamento de Dependências

- Python: Usa ferramentas como pip e venv para gerenciar dependências e ambientes virtuais.

- Java: Usa ferramentas como Maven ou Gradle para gerenciar dependências e construir projetos.

#Abstração
Abstração refere-se ao processo de ocultar os detalhes complexos de uma implementação e expor apenas as funcionalidades essenciais. Exemplo: Car encapsula detalhes internos e apenas expõem de funcionalidades essenciais como  `start_engine`, `stop_engine` e `drive`.

````python
#Vamos criar uma classe Car que abstrai o funcionamento interno de um carro.
class Car:
    """Representa um carro."""

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self._engine_started = False

    def start_engine(self):
        """Inicia o motor do carro."""
        self._engine_started = True
        print(f"O motor do {self.model} foi iniciado.")

    def stop_engine(self):
        """Desliga o motor do carro."""
        self._engine_started = False
        print(f"O motor do {self.model} foi desligado.")

    def drive(self):
        """Dirige o carro."""
        if self._engine_started:
            print(f"O {self.model} está dirigindo.")
        else:
            print(f"Inicie o motor do {self.model} primeiro.")
````

#Testando!

````python
car = Car('Toyota', 'Corolla', 2020)
car.start_engine()  # O motor do Corolla foi iniciado.
car.drive()         # O Corolla está dirigindo.
car.stop_engine()   # O motor do Corolla foi desligado.
````

# Abstração em Python vs Java: Diferenças e Similaridades

## Similaridades
### Encapsulamento de Detalhes Internos

- Tanto Python quanto Java permitem encapsular detalhes internos de uma classe, expondo apenas o que é necessário para o usuário.

- Em ambas as linguagens, você pode definir atributos e métodos privados (ou protegidos) para esconder a implementação interna.

### Uso de Métodos Públicos para Interação

- Em ambas as linguagens, a interação com um objeto é feita por meio de métodos públicos, que abstraem a complexidade interna.

- Por exemplo, em Python:
```python
car.start_engine()
```
- Em Java:
```java
car.ligarMotor();
```

### Foco no "O Que" em vez do "Como"

- Tanto Python quanto Java incentivam o uso de abstração para focar no o que um objeto faz, em vez de como ele faz. Isso facilita a manutenção e a reutilização do código

##  Diferenças
### Modificadores de Acesso

- Java: Java possui modificadores de acesso explícitos, como private, protected e public, que controlam rigorosamente o acesso a atributos e métodos.

- Exemplo:
```java   
private boolean motorLigado; // Atributo privado
public void ligarMotor() { } // Método público
```

- Python:

- Python não possui modificadores de acesso tão rígidos. Por convenção, usa-se um sublinhado (_) para indicar que um atributo ou método é privado, mas isso não impede o acesso direto.

- Exemplo:
```python
self._engine_started = False  # Atributo "privado" por convenção
def start_engine(self): pass  # Método público
```

### Abstração com Classes Abstratas e Interfaces

- Java:

    - Java possui suporte nativo para classes abstratas e interfaces, que são ferramentas poderosas para implementar abstração.

    - Exemplo de classe abstrata:
     ```java
      public abstract class Veiculo {
    public abstract void mover();
}
     ```
    - Exemplo de interface:
     ```java
       public interface Ligavel {
       void ligar();
    void desligar();
}
     ```

- Python:

    - Python também suporta classes abstratas, mas por meio do módulo abc (Abstract Base Classes). No entanto, não possui interfaces nativas.

    - Exemplo de classe abstrata:
    ```python
    from abc import ABC, abstractmethod

    class Veiculo(ABC):
        @abstractmethod
        def mover(self):
            pass
      ```

| Característica               | Python                                  | Java                                    |
|------------------------------|-----------------------------------------|-----------------------------------------|
| **Modificadores de Acesso**  | Convenção (`_` para privado)            | Explícitos (`private`, `public`, etc.)  |
| **Classes Abstratas**        | Suportado via módulo `abc`              | Suporte nativo                          |
| **Interfaces**               | Não possui interfaces nativas           | Suporte nativo                          |
| **Filosofia**                | Flexível e dinâmica                     | Rigorosa e explícita                    |
| **Foco**                     | Simplicidade e convenção                | Segurança e clareza                     |

#Encapsulamento

Encapsulamento refere-se à prática de restringir o acesso direto a alguns dos componentes de um objeto e usar métodos para modificar esses componentes.

````python
#Vamos modificar o exemplo do carro para encapsular o atributo _engine_started.

class Car:
    """Representa um carro."""

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.__engine_started = False  # Atributo privado

    def start_engine(self):
        """Inicia o motor do carro."""
        self.__engine_started = True
        print(f"O motor do {self.model} foi iniciado.")

    def stop_engine(self):
        """Desliga o motor do carro."""
        self.__engine_started = False
        print(f"O motor do {self.model} foi desligado.")

    def drive(self):
        """Dirige o carro."""
        if self.__engine_started:
            print(f"O {self.model} está dirigindo.")
        else:
            print(f"Inicie o motor do {self.model} primeiro.")

    def is_engine_started(self):
        """Verifica se o motor está iniciado."""
        return self.__engine_started
````

#Testando!

````python
car = Car('Toyota', 'Corolla', 2020)
print(car.is_engine_started())  # False
car.start_engine()              # O motor do Corolla foi iniciado.
print(car.is_engine_started())  # True
car.drive()                     # O Corolla está dirigindo.
car.stop_engine()               # O motor do Corolla foi desligado.
print(car.is_engine_started())  # False
````

# Encapsulamento em Python vs Java: Diferenças e Similaridades

#### O encapsulamento é um conceito fundamental da programação orientada a objetos (POO) que visa proteger os dados internos de um objeto, expondo apenas uma interface controlada para interagir com ele.

### Similaridades

#### Uso de Métodos Públicos para Acesso Controlado:

- Tanto Python quanto Java usam métodos públicos (como getters e setters) para permitir o acesso controlado aos atributos privados de uma classe.

- Exemplo:

            Em Java: `public int getValor() { return this.valor; }`

            Em Python: `def get_valor(self): return self.__valor`

#### Proteção de Dados Internos:

- Ambas as linguagens permitem ocultar os detalhes internos de uma classe, expondo apenas o necessário para o uso externo. Isso ajuda a evitar que os dados sejam modificados de maneira incorreta ou inconsistente.

#### Foco na Interface Pública:

- Em ambas as linguagens, a ideia é que o usuário da classe interaja apenas com a interface pública (métodos públicos), sem se preocupar com a implementação interna.

### Diferenças

#### Modificadores de Acesso

- **Java**: Possui modificadores de acesso explícitos, como `private`, `protected` e `public`, que definem claramente o nível de acesso aos atributos e métodos.
  - Exemplo: `private int valor;` (só pode ser acessado dentro da classe).

- **Python**: Não possui modificadores de acesso explícitos. Em vez disso, usa convenções de nomenclatura para indicar a visibilidade:
  - Um underscore (`_`) indica que um atributo ou método é "protegido" (não deve ser acessado externamente, mas não é estritamente privado).
  - Dois underscores (`__`) indicam que um atributo ou método é "privado" (o nome é alterado internamente para evitar acesso direto).

#### Acesso a Atributos Privados

- **Java**: Atributos privados não podem ser acessados diretamente de fora da classe. O acesso só é possível através de métodos públicos (getters e setters).
  - Exemplo: `obj.getValor()` (não é possível acessar `obj.valor` diretamente).

- **Python**: Atributos privados (com `__`) ainda podem ser acessados de fora da classe, mas o nome do atributo é alterado internamente (name mangling). Isso torna o acesso mais difícil, mas não impossível.
  - Exemplo: Um atributo `__valor` pode ser acessado como `_Classe__valor`.

#### Filosofia de Design

- **Java**: Segue uma abordagem mais rígida e explícita, com regras claras sobre o que pode ou não ser acessado. O encapsulamento é fortemente reforçado pela linguagem.

- **Python**: Segue uma filosofia mais flexível, baseada na confiança no desenvolvedor. A linguagem assume que o programador sabe o que está fazendo e não impõe restrições rígidas.

#### Convenções e Boas Práticas:

- **Java**: A linguagem incentiva o uso de encapsulamento estrito, com atributos privados e métodos públicos para interagir com eles.

- **Python**: A linguagem é mais flexível e incentiva o uso de encapsulamento apenas quando necessário, seguindo a filosofia "We are all consenting adults here" (todos somos adultos consentidos).

#Definição de classe em Python
Aqui está um exemplo básico de uma classe CreditCard com um construtor \_\_init\_\_ que inicializa as variáveis de instância.

````python
class CreditCard:
    """
    Uma classe para representar um cartão de crédito.

    Attributes:
    customer (str): O nome do cliente.
    bank (str): O nome do banco.
    account (str): O número da conta.
    limit (float): O limite de crédito.
    balance (float): O saldo atual (padrão é 0).
    """

    def __init__(self, customer, bank, account, limit):
        """
        Inicializa um objeto CreditCard com os dados fornecidos.

        Parameters:
        customer (str): O nome do cliente.
        bank (str): O nome do banco.
        account (str): O número da conta.
        limit (float): O limite de crédito.
        """
        self._customer = customer
        self._bank = bank
        self._account = account
        self._limit = limit
        self._balance = 0

    def get_customer(self):
        """
        Retorna o nome do cliente.

        Returns:
        str: O nome do cliente.
        """
        return self._customer

    def get_bank(self):
        """
        Retorna o nome do banco.

        Returns:
        str: O nome do banco.
        """
        return self._bank

    def get_account(self):
        """
        Retorna o número da conta.

        Returns:
        str: O número da conta.
        """
        return self._account

    def get_limit(self):
        """
        Retorna o limite de crédito.

        Returns:
        float: O limite de crédito.
        """
        return self._limit

    def get_balance(self):
        """
        Retorna o saldo atual.

        Returns:
        float: O saldo atual.
        """
        return self._balance

        #Métodos Adicionais--> Vamos adicionar os métodos charge e make_payment à classe CreditCard.

    def charge(self, price):
        """
        Adiciona uma cobrança ao saldo do cartão de crédito.

        Parameters:
        price (float): O valor da cobrança.

        Returns:
        bool: True se a cobrança for bem-sucedida (dentro do limite), False caso contrário.
        """
        if price + self._balance > self._limit:
            return False
        else:
            self._balance += price
            return True

    def make_payment(self, amount):
        """
        Faz um pagamento, reduzindo o saldo do cartão de crédito.

        Parameters:
        amount (float): O valor do pagamento.
        """
        self._balance -= amount




````
